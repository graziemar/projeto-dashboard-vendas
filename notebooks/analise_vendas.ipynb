```python
import pandas as pd
import random
!pip install faker
from faker import Faker
from datetime import datetime, timedelta

fake = Faker('pt_BR')
```

```python
produtos = [
    'Notebook',
    'Mouse',
    'Teclado',
    'Monitor',
    'Headset',
    'Cadeira Gamer',
    'Webcam',
    'Tablet'
]

categorias = {
    'Notebook': 'Informática',
    'Mouse': 'Periféricos',
    'Teclado': 'Periféricos',
    'Monitor': 'Informática',
    'Headset': 'Áudio',
    'Cadeira Gamer': 'Móveis',
    'Webcam': 'Acessórios',
    'Tablet': 'Mobile'
}

vendedores = [
    'Ana',
    'Carlos',
    'Julia',
    'Marcos',
    'Fernanda'
]

regioes = [
    'Sudeste',
    'Sul',
    'Nordeste',
    'Centro-Oeste',
    'Norte'
]
```

```python
dados = []

for i in range(1, 301):

    produto = random.choice(produtos)

    quantidade = random.randint(1,5)

    preco = random.randint(100,5000)

    faturamento = quantidade * preco

    data_venda = fake.date_between(
        start_date='-12M',
        end_date='today'
    )

    dados.append({
        'id_venda': i,
        'data_venda': data_venda,
        'cliente': fake.company(),
        'produto': produto,
        'categoria': categorias[produto],
        'vendedor': random.choice(vendedores),
        'regiao': random.choice(regioes),
        'quantidade': quantidade,
        'preco_unitario': preco,
        'faturamento': faturamento,
        'meta_batida': random.choice(['Sim', 'Não'])
    })
```

```python
df_vendas = pd.DataFrame(dados)

df_vendas.head()
```

```python
df_vendas.to_csv(
    'base_vendas.csv',
    index=False
)
```
